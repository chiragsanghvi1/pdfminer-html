```python
import io

from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
```

```python
from pdfminer.high_level import extract_text
import re

def extract_headings(pdf_path, threshold=2):
    headings = []
    current_heading = ""
    heading_level = None
    heading_pattern = re.compile(r"^(#+)\s*(.*)$")

    def process_heading(heading, level):
        # Modify this function to decide what to do with the extracted headings.
        # For now, we'll just print them.
        print(f"Level {level}: {heading}")

    def check_heading(line):
        nonlocal current_heading, heading_level
        match = heading_pattern.match(line)
        if match:
            level = len(match.group(1))
            heading_text = match.group(2).strip()
            if heading_level is None or level > heading_level:
                current_heading = heading_text
                heading_level = level
            elif level <= heading_level:
                process_heading(current_heading, heading_level)
                current_heading = heading_text
                heading_level = level

    text = extract_text(pdf_path)

    for line in text.splitlines():
        check_heading(line)

    # Process the last heading encountered in the PDF
    if current_heading:
        process_heading(current_heading, heading_level)
```

```python
pdf_path = "example.pdf"
extract_headings(pdf_path)
```
